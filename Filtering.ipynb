```julia
using ImageFiltering, Statistics, FFTW, Plots, DSP, PlutoUI, TestImages, ImageView
```

```julia
function show_coloured_kernel(kernel)
    to_rgb(x) = RGB(max(-x,0), max(x, 0), 0)
    to_rgb.(kernel)/maximum(abs.(kernel))
end
```

```julia
kernel = Kernel.gaussian((1,1))
show_coloured_kernel(kernel)
```

```julia
img = testimage("mandrill");
imgg = imfilter(img, Kernel.gaussian(3));
imgl = imfilter(img, Kernel.Laplacian());
[ImageView.imshow(img), ImageView.imshow(imgg), ImageView.imshow(imgl)]
```
